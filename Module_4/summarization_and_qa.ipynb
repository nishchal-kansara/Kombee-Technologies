```python
from transformers import pipeline
```

```python
summarizer = pipeline(
    task="summarization",
    model="facebook/bart-large-cnn"
)
```

```python
text = """
Diwali is one of the most important festivals celebrated in India.
It is known as the festival of lights and symbolizes the victory of light over darkness.
People decorate their homes with lamps, burst fireworks, and share sweets with family and friends.
The festival also has deep cultural and religious significance across different regions of India.
"""
```

```python
summary = summarizer(
    text,
    max_length=60,
    min_length=25
)

print("Summary")
print(summary[0]["summary_text"])
```

```python
qa_model = pipeline(
    task="question-answering",
    model="distilbert-base-cased-distilled-squad"
)
```

```python
context = text
```

```python
questions = [
    "What is Diwali known as",
    "Why is Diwali celebrated",
    "How do people celebrate Diwali"
]
```

```python
for q in questions:
    answer = qa_model(
        question=q,
        context=context
    )
    print("Question")
    print(q)
    print("Answer")
    print(answer["answer"])
    print()
```
