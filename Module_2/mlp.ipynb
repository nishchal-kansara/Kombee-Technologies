```python
import torch
import math
```

```python
# create dummy dataset
x = torch.linspace(-5, 5, 100).reshape(-1, 1)
y = 2 * x + 1
```

```python
# Initialize Parameters
torch.manual_seed(0)

W1 = torch.randn(1, 10)
b1 = torch.zeros(10)

W2 = torch.randn(10, 1)
b2 = torch.zeros(1)
```

```python
# Activation Funtion
def relu(z):
    return torch.clamp(z, min=0)

def relu_grad(z):
    return (z > 0).float()
```

```python
# Forward Pass
def forward(x):
    z1 = x @ W1 + b1
    a1 = relu(z1)
    z2 = a1 @ W2 + b2
    return z1, a1, z2
```

```python
# Mean Squared Error
def mse_loss(y_pred, y_true):
    return ((y_pred - y_true) ** 2).mean()
```

```python
# Backward Pass
def backward(x, y, z1, a1, y_pred):
    global W1, b1, W2, b2

    n = x.shape[0]

    dloss_dy = 2 * (y_pred - y) / n

    dW2 = a1.t() @ dloss_dy
    db2 = dloss_dy.sum(0)

    da1 = dloss_dy @ W2.t()
    dz1 = da1 * relu_grad(z1)

    dW1 = x.t() @ dz1
    db1 = dz1.sum(0)

    return dW1, db1, dW2, db2
```

```python
# Training
lr = 0.01
epochs = 500

for epoch in range(epochs):
    z1, a1, y_pred = forward(x)
    loss = mse_loss(y_pred, y)

    dW1, db1, dW2, db2 = backward(x, y, z1, a1, y_pred)

    W1 -= lr * dW1
    b1 -= lr * db1
    W2 -= lr * dW2
    b2 -= lr * db2

    if epoch % 50 == 0:
        print("Epoch", epoch, "Loss", loss.item())
```

```python
# Testing
test_x = torch.tensor([[4.0]])
_, _, pred = forward(test_x)
print("Prediction", pred.item())
```
