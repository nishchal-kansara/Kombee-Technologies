```python
with open("diwali.txt", "r", encoding="utf-8") as f:
    raw_text = f.read()

print(raw_text[:500])
```

```python
import re

clean_text = raw_text.lower()
clean_text = re.sub(r"\s+", " ", clean_text)

print(clean_text[:500])
```

```python
sentences = re.split(r"[.!?]", clean_text)
sentences = [s.strip() for s in sentences if s.strip()]

for s in sentences[:3]:
    print(s)
```

```python
words = clean_text.split()
print(words[:20])
print("Total words", len(words))
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
sample_sentence = "Diwali is a festival of lights"

tokens = tokenizer.tokenize(sample_sentence.lower())
token_ids = tokenizer.convert_tokens_to_ids(tokens)

print("Tokens")
print(tokens)

print("Token IDs")
print(token_ids)
```

```python
encoded = tokenizer(
    clean_text,
    add_special_tokens=True,
    truncation=False
)

print("Number of tokens", len(encoded["input_ids"]))
print(encoded["input_ids"][:20])
```

```python
tokens = tokenizer.convert_ids_to_tokens(encoded["input_ids"][:30])
print(tokens)
```
