```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(0, 10)
y = x ** 2

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="cos(x)")

plt.title("Trigonometric Functions")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
x = np.linspace(-10, 10, 200)
y = x ** 3

plt.plot(x, y)
plt.title("y = x^3")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
np.random.seed(0)
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.title("Random Scatter")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x = np.random.rand(100)
y = x ** 2

plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("x")
plt.show()
```

```python
categories = ["A", "B", "C", "D"]
values = [23, 45, 12, 36]

plt.bar(categories, values)
plt.title("Category Values")
plt.ylabel("Value")
plt.show()
```

```python
plt.barh(categories, values)
plt.show()
```

```python
data = np.random.randn(1000)

plt.hist(data, bins=30)
plt.title("Histogram of Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
data = np.random.randn(10_000)
plt.hist(data, bins=50)
plt.show()
```

```python
x = np.linspace(0, 5, 100)

plt.figure()
plt.plot(x, x)
plt.title("Linear")

plt.figure()
plt.plot(x, x ** 2)
plt.title("Quadratic")

plt.show()
```

```python
plt.plot(x, x ** 2)
plt.title("Saved Plot")
plt.savefig("quadratic_plot.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
df = pd.DataFrame({
"x": np.arange(10),
"y": np.random.randint(1, 20, size=10)
})

df.plot(x="x", y="y")
plt.show()
```
